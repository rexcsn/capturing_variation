```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap, supported_projections
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import seaborn as sb

import json
from gensim.models import Doc2Vec
from sklearn.cluster import AgglomerativeClustering
from matplotlib import colors
from shapely.geometry import Point, shape

sb.set_context('poster')

cm = plt.get_cmap('gist_rainbow')

```

```python

def get_shortest_in(needle, haystack):
    '''
    :param needle: single (lat,long) tuple.
    :param haystack: numpy array to find the point in that has the shortest distance to needle
    :return:
    '''
    dlat = np.radians(haystack[0]) - np.radians(needle[0])
    dlon = np.radians(haystack[1]) - np.radians(needle[1])
    a = np.square(np.sin(dlat / 2.0)) + np.cos(np.radians(needle[0])) * np.cos(np.radians(haystack[0])) * np.square(
        np.sin(dlon / 2.0))
    great_circle_distance = 2 * np.arcsin(np.minimum(np.sqrt(a), np.repeat(1, 1)))
    d = 6371 * great_circle_distance
    return d.tolist()[0]
```

```python
location_file = '/Users/dirkhovy/Dropbox/Computational Sociolinguistics/jodel/data/country_stuff.json'
corpus = '/Users/dirkhovy/Dropbox/Computational Sociolinguistics/jodel/models/no_cities.citycounts'
model = '/Users/dirkhovy/Dropbox/Computational Sociolinguistics/jodel/models/no_cities.model'
clusters = 11
mmin = 200
mmax = 9999999999
restrict = 'DE'

# read in locations
locations = {}
location_list = json.load(open(location_file))
for location in location_list:
    try:
        if restrict is None or restrict == location['country']:
            locations[location['city']] = [(location['lat'], location['lng']), location['country']]
    except KeyError:
        if restrict is None or restrict == location['country']:
            locations[location['location']] = [(location['lat'], location['lng']), location['country']]

with open(corpus, encoding='utf-8') as corpus_file:
    city_density = json.load(corpus_file)

lower_bound = min(city_density.values())
upper_bound = max(city_density.values())
divisor = upper_bound - lower_bound

# scale counts
city_density_scaled = {}
for city in city_density:
    city_density_scaled[city] = ((city_density[city] - lower_bound) / divisor) * 500 + 50

model = Doc2Vec.load(model)
eligible_cities = [city for city in locations if city in city_density if city_density[city] >= mmin and city_density[city] <= mmax and city in model.docvecs]

print('\n\tusing %s cities' % len(eligible_cities))

color_names = ['red' for city in eligible_cities]
cMap = 'Reds'

# get clusters
proximity = np.zeros((len(eligible_cities), len(eligible_cities)))
for i, city1 in enumerate(eligible_cities):
    for j in range(i+1, len(eligible_cities)):
        city2 = eligible_cities[j]
        distance = get_shortest_in(locations[city1][0], locations[city2][0])

        proximity[i, j] = distance
        proximity[j, i] = distance

# invert distances
furthest = proximity.max()
proximity = furthest - proximity

cluster_colors = [cm(1. * i / clusters) for i in range(clusters)]

vectors = np.array([model.docvecs[city] for city in eligible_cities])
```

```python
clustering_algo = AgglomerativeClustering(n_clusters=clusters, connectivity=proximity)
cluster_ids = clustering_algo.fit_predict(X=vectors)
color_names = [cluster_colors[c] for i, c in enumerate(cluster_ids) if locations[eligible_cities[i]][-1] == "DE"]
cMap = colors.ListedColormap(cluster_colors)

```

```python
fig, ax = plt.subplots(figsize=(10.4, 14))

m = Basemap(llcrnrlon=6.15,
            urcrnrlon=15.49, 
            urcrnrlat=54.925,
            llcrnrlat=47.2, 
            lat_ts=50,lat_0=50,lon_0=9.5,
            resolution='i', projection='gnom', epsg=4326)
# background = plt.imread('figure1_Lameli.png')
# m.imshow(background, interpolation='None', origin='upper', zorder=1)
# m.drawcoastlines(color='red')
# m.drawcountries(zorder=5, color='red')
# x, y = m([8.268634], [49.051938], inverse=False)
# m.plot(x,y, zorder=6, marker='o',color='Red')

x, y, z = zip(*[(locations[city][0][1], locations[city][0][0], city_density_scaled[city]) for city in eligible_cities if locations[city][-1] == "DE"])

# transform coordinates into map space
x1, y1 = m(x, y)
m.scatter(x1, y1, z, marker='o', c=color_names, cmap=cMap, zorder=4, alpha=0.8)

# m.readshapefile('/Users/dirkhovy/Dropbox/working/lowlands/sociolinguistics/playground/Lameli maps/Lameli', 'de', drawbounds=True)
m.readshapefile('/Users/dirkhovy/Dropbox/working/lowlands/sociolinguistics/playground/Lameli maps/org_new/SIM', 'de', drawbounds=True)

patches = []

for info, dshape in zip(m.de_info, m.de):
    poly = Polygon(np.array(dshape, dtype=np.float), True)
    patches.append(poly)

#         x, y = zip(*shape) 
#         m.plot(x, y, marker=None, color='r')
               
pc = PatchCollection(patches, facecolor= 'grey', edgecolor='k', linewidths=.3, zorder=2)
pc.set_alpha(0.5)
ax.add_collection(pc)

lons, lats= zip(*[coords for city, (coords, iso) in locations.items() if city in eligible_cities and locations[city][-1] == "DE"])
lats1, lons1 = m(list(lats), list(lons))

for name, xpt, ypt in zip(eligible_cities, lats1, lons1):
    if  name == 'Wrzburg':#city_density_scaled[name] >= 250 or name in {'Basel', 'Zrich', 'Bern', 'Genf', 'Trier', 'Saarbrcken', 'Lausanne'}:
        plt.text(xpt, ypt, name, fontsize="16", zorder=5)

addon = '_DE' if restrict == 'DE' else ''

fig.savefig('/Users/dirkhovy/Dropbox/Computational Sociolinguistics/jodel/_results/maps/no_cities/lameli/%sclusters%s.png' % (clusters, addon), dpi=200, bbox_inches='tight', pad_inches=0.1)   
plt.show()
```

```python
targets = {'Frankfurt am Main', 'Berlin', 'Hamburg', 'Mnchen', 'Stuttgart', 'Dortmund'}

X_train = []
y_train = []
for line in open('../../jodel/models/no_cities.corpus'):
    line = json.loads(line)
    if line['tags'][0] in targets:
        X_train.append(model.infer_vector(line['words']))
        y_train.append(line['tags'][0])
    if len(X_train) == 100000:
        break
        
        
X_test = []
y_test = []
for line in open('../../jodel/models/no_cities.corpus'):
    line = json.loads(line)
    if line['tags'][0] in targets:
        X_test.append(model.infer_vector(line['words']))
        y_test.append(line['tags'][0])
    if len(X_test) == 10000:
        break
        
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

clf = LogisticRegression()
clf.fit(X_train, y_train)

from sklearn.metrics import f1_score
f1_score(y_test, clf.predict(X_test), average='micro')
```
